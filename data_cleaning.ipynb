### ***Data Cleaning***

*i) Filter out useful Features*

```python
data.drop(columns=[
    'listing_url','scrape_id','picture_url','host_id','host_picture_url',
    'host_thumbnail_url','host_neighbourhood','host_listings_count',
    'neighbourhood_group_cleansed','minimum_nights','maximum_nights',
    'minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights',
    'maximum_maximum_nights', 'minimum_nights_avg_ntm','maximum_nights_avg_ntm',
    'calendar_updated','calendar_last_scraped','number_of_reviews_ltm',
    'review_scores_accuracy','review_scores_value','calculated_host_listings_count',
    'calculated_host_listings_count_entire_homes','calculated_host_listings_count_private_rooms',
    'calculated_host_listings_count_shared_rooms','reviews_per_month'
    ],axis=1,inplace=True)
```

*ii) Rename the filtered Features*

```python
data.rename(columns={
    'last_scraped':'current_date',
    'neighborhood_overview':'neighbourhood_overview'
    },inplace=True)
```

*iii) Modify DataTypes*

*<font color='lightgreen'>- Convert current_date, host_since, first_review and last_reviews from object to datetime</font>*

```python
dates=['current_date','host_since','first_review','last_review']
for index in dates:
    data[index]=pd.to_datetime(data[index])
```

*<font color='lightgreen'>- Convert bathrooms_text from object to integer*

*Convert specific texts to numbers*

```python
text=['half','Half']
for index in text:
    data['bathrooms_text']=np.where(data['bathrooms_text'].str.contains(index),1,data['bathrooms_text'])
```

*Split the Feature bathrooms_text*

```python
data[['bathrooms','text','text1']]=data['bathrooms_text'].str.split(expand=True,n=2)
data.drop(['text','text1'],axis=1,inplace=True)
```

*Convert the Numeric values to integers*

```python
data['bathrooms']=round(data['bathrooms'].astype(float))
data['bathrooms']=data['bathrooms'].astype('Int64')
```

*<font color='lightgreen'>- Convert beds and bedrooms from float to integers</font>*

```python
data['beds']=data['beds'].astype('Int64')
data['bedrooms']=data['bedrooms'].astype('Int64')
```

*<font color='lightgreen'>- Remove the square brackets and apostrophe in the Features amenities and verifications</font>*

```python
data['amenities']=data['amenities'].str.replace('[','').str.replace(']','').str.replace('"','')
data['host_verifications']=data['host_verifications'].str.replace('[','').str.replace(']','').str.replace("'","")
```

*<font color='lightgreen'>- Convert Price from object to integer</font>*

```python
data['price']=data['price'].str.replace('$','').str.replace(',','')
data['price']=pd.to_numeric(data['price']).astype('Int64')
```

*<font color='lightgreen'>- Replace Null/NA values with Not Mentioned for object datatypes</font>*

```python
data['host_location']=np.where(data['host_location'].str.contains('Africa'),data['host_location'].str.strip(),'Unknown')
data['neighbourhood']=np.where(data['neighbourhood'].str.contains('Africa'),data['neighbourhood'].str.strip(),'Unknown')
```

```python
cols=[
    'description','neighbourhood_overview','host_about','host_response_time',
    'host_response_rate','host_acceptance_rate','host_is_superhost',
    'host_location','neighbourhood','license'
    ]

for index in cols:
    data.replace({
        index:np.NAN
    },'Not Mentioned',inplace=True)
```

*<font color='lightgreen'>- Filling Null Values in Numeric Columns with backfill</font>*

```python
cols=['bathrooms','beds','bedrooms']
for index in cols:
    data[index].fillna(method='bfill',inplace=True)
```

```python
col=['review_scores_rating', 'review_scores_cleanliness','review_scores_checkin', 'review_scores_communication','review_scores_location']
for index in col:
    data[index].fillna(method='bfill',inplace=True)
```

*<font color='lightgreen'>- Fill Null Values in the Date Columns using Interpolate</font>*

```python
date=['first_review','last_review']
for index in date:
    data[index]=data[index].interpolate().dt.date
    data[index]=pd.to_datetime(data[index])
    
```

*Preprocessed Data*

```python
data=data[[
    'id','source','name','description','neighbourhood_overview',
    'host_url','host_name','host_about','host_since','current_date','host_response_time','host_response_rate',
    'host_acceptance_rate','host_is_superhost','host_total_listings_count',
    'host_verifications','host_has_profile_pic','host_identity_verified',
    'host_location','latitude','longitude','neighbourhood','neighbourhood_cleansed','property_type',
    'room_type','license','instant_bookable','accommodates','bedrooms','beds','bathrooms','amenities','price','has_availability',
    'availability_30','availability_60','availability_90','availability_365','number_of_reviews',
    'number_of_reviews_l30d','first_review','last_review','review_scores_rating','review_scores_cleanliness',
    'review_scores_checkin','review_scores_communication','review_scores_location'
    ]].copy()
```
